# Producing our merger sample

```python
# producting an array with objects used in RF + all object with p_RF=None which were excluded in RF
temp = np.array(merged_68.copy(), dtype=None)
rf_objects = []
for i in range(len(temp)):
    rf_objects.append([temp[i,1],temp[i,4]])
rf_objects = np.stack(rf_objects)

#object IDs, which are needed to find these objects in the image files
subid, indices = np.unique(rf_objects[:,0], return_index=True)
RF_labels = rf_objects[indices]
```

```python
shid_array = RF_labels[:,0]
merger_labels = RF_labels[:,1]
```

```python
def files(param_name, type_name):
    name = merged_68[f'{param_name}'].values
    np.save(dir_test + f'{type_name}.npy', name)

files('merger', 'subsamples_y')
files('subid', 'subsamples_ID')
files('cc_I', 'concentration')
files('m20_I', 'm20')
files('logmstar', 'logMstar')
```

```python
list_of_mergers=[]
list_of_nonmergers=[]
```

```python
#subdirs for displaying
path = "C:/Users/aleje/OneDrive - Curtin/Uni Y3/ICRAR Intern/deepmerge_files/"
subdirs = ['res_subdir_{josh:03d}'.format(josh=x) for x in range(23)]
```

```python
#All 4 filters i order ['ACS-F435W', ACS-F814W','NC-F356W','WFC3-F160W'], we will need second and fourth:
for shid, merger in zip(shid_array,merger_labels):
    for camnum in ['00','01','02','03']:
            filter_files=np.sort(np.asarray(glob.glob(os.path.join(path,'res_sub*','*sh'+str(shid)+'cam'+camnum+'*SB25.npy'))))
            # the above should return exactly 4 files
            
            if filter_files.shape[0] != 4:
                continue

            tmp= np.zeros((4, 75, 75))

            tmp[0,:,:]=np.load(filter_files[0])
            tmp[1,:,:]=np.load(filter_files[1])
            tmp[2,:,:]=np.load(filter_files[2])
            tmp[3,:,:]=np.load(filter_files[3])
     

            if bool(merger) is True:
               list_of_mergers.append(tmp)

            else:
                list_of_nonmergers.append(tmp)
                
```

```python
mergers = np.stack(list_of_mergers)
notmergers = np.stack(list_of_nonmergers)
mergers_y = np.ones(mergers.shape[0])
notmergers_y = np.zeros(notmergers.shape[0])
```

```python
X = np.vstack((mergers, notmergers))
y = np.concatenate((mergers_y, notmergers_y))
```

```python
# producing numpy arrays with all objects in 4 filters
np.save('SB25.npy', X)
np.save('SB25_y.npy', y)
```

# Merger augmentation, plus extracton of only 2 relevant filters

```python
XX = np.load('SB25.npy')
yy = np.load('SB25_y.npy')
```

```python
xx_data_list = []
for i in range(len(X)):
    xx_data_list.append([X[i,1],X[i,3]])
xx_data = np.stack(xx_data_list)
yy_data = y
```

```python
list_ud = []
list_lr = []
list_rot = []
list_rot180 = []

for i in range(0,1624):
    tmp = np.zeros((2, 75, 75))
    tmp[0,:,:]=np.flipud(xx_data[i,0])
    tmp[1,:,:]=np.flipud(xx_data[i,1])
    list_ud.append(tmp)
    
    
for i in range(0,1624):
    tmp = np.zeros((2, 75, 75))
    tmp[0,:,:]=np.fliplr(xx_data[i,0])
    tmp[1,:,:]=np.fliplr(xx_data[i,1])
    list_lr.append(tmp)
    
    
for i in range(0,1624):
    tmp = np.zeros((2, 75, 75))
    tmp[0,:,:]=np.rot90(xx_data[i,0])
    tmp[1,:,:]=np.rot90(xx_data[i,1])
    list_rot.append(tmp)
    
    
for i in range(0,1624):
    tmp = np.zeros((2, 75, 75))
    tmp[0,:,:]=np.rot90(np.rot90(xx_data[i,0]))
    tmp[1,:,:]=np.rot90(np.rot90(xx_data[i,1]))
    list_rot180.append(tmp)
```

```python
mergers_ud = np.stack(list_ud)
mergers_lr = np.stack(list_lr)
mergers_rot = np.stack(list_rot)
mergers_rot180 = np.stack(list_rot180)
```

```python
y_ud = np.ones(mergers_ud.shape[0])
y_lr = np.ones(mergers_lr.shape[0])
y_rot = np.ones(mergers_rot.shape[0])
y_rot180 = np.ones(mergers_rot180.shape[0])
```

```python
X_augmented = np.vstack((xx_data[:1624], mergers_ud, mergers_lr, mergers_rot, mergers_rot180, xx_data[1624:]))
y_augmented = np.concatenate((yy_data[:1624], y_ud, y_lr, y_rot, y_rot180, yy_data[1624:]))
```

```python
np.save('SB25_augmented.npy', X_augmented)
np.save('SB25_augmented_y.npy', y_augmented)
```
